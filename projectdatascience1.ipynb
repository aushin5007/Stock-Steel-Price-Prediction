```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
st=pd.read_csv(r"/content/project data.csv")
print(st.head)
df=pd.DataFrame(st)
print(df)
print(df['Date'])
print(df['Tata Steel'])
plt.figure(figsize=(8,4))
plt.plot(df['Date'],df['Tata Steel'])
plt.title("Tata Steel")
plt.xlabel('Month')
plt.ylabel('Stock price')
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['Date'],df['Jindal Steel'])
plt.title("Jindal Steel")
plt.xlabel('Month')
plt.ylabel('Stock price')
plt.grid(True)
plt.show()
```
